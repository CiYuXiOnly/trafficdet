```python
from sklearn.svm import SVC
import numpy as np
import pandas as pd
```

```python
# 创建示例DataFrame对象
df1 = pd.read_csv("data/featured_csv/benign_small_flow.csv")
df2 = pd.read_csv("data/featured_csv/malicious_small_flow.csv")
df = pd.concat([df1, df2])
df = df.drop(["src", "sport", "dst", "dport"], axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df.values)
# 将标准化后的数据转换回 DataFrame
df_scaled = pd.DataFrame(data_scaled, columns=df.columns)
df_scaled
```

```python
# 将DataFrame转换为NumPy数组
data = df.values

# 设置批次大小
batch_size = 2

# 创建SVC分类器并进行批量训练
svm = SVC(kernel='linear', C=1.0, random_state=42)

# 分批次进行训练
for i in range(0, len(data), batch_size):
    batch_data = data[i:i+batch_size]
    X_batch = batch_data[:, :-1]  # 输入特征
    y_batch = batch_data[:, -1]  # 标签
    svm.fit(X_batch, y_batch)

# 在验证集上进行预测
X_test = np.array([[3, 8], [4, 9]])
y_pred = svm.predict(X_test)

print(y_pred)  # 预测结果
```
